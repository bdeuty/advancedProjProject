```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
data_filepath = "TableS8_T2T_KrakenUniq_BIO_Fullset.csv"
metadata_filepath = "TableS9_metadata_KrakenUniq_BIO_Fullset.csv"

data = pd.read_csv(data_filepath, index_col=0)
metadata = pd.read_csv(metadata_filepath)
metadata = metadata[['sampleid', 'vital_status']]
metadata['vital_binary'] = metadata['vital_status'].apply(lambda x: 1 if x == "Alive" else 0)
metadata = metadata[['sampleid', 'vital_binary']]

sample_labels = metadata.set_index('sampleid')
labels = sample_labels.loc[data.index, 'vital_binary'].values
print("Data:\n")
print(data.iloc[:5,:5])
print("\nMetadata:\n")
print(metadata.head(5))
print("\nLabels:\n")
print(labels[:10])
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

X_train, X_test, y_train, y_test = train_test_split(data_scaled, labels,test_size=0.2,random_state=42)
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

y_pred_proba = clf.predict_proba(X_test)[:,1]

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=1, label="Random Guess")
plt.xlabel("False Positive Rate (1 - Specificity)")
plt.ylabel("True Positive Rate (Sensitivity)")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()
#plt.savefig('ROC_curve.png')
```

```python
location_data = pd.read_csv(metadata_filepath)
location_data = location_data[['data_submitting_center_label']]  # Assuming this is the column name
location_data['data_submitting_center_label'].fillna("Unknown", inplace=True)

# Encode the location data (One-Hot Encoding)
encoder = OneHotEncoder(sparse=False)
location_encoded = encoder.fit_transform(location_data.values.reshape(-1, 1))

# Create the contingency table (location vs vital status)
location_status_data = pd.crosstab(location_data['data_submitting_center_label'], labels)

# Perform the Chi-square test for independence
chi2, p_value, _, _ = chi2_contingency(location_status_data)
print(f"Chi-square Test p-value: {p_value}")

# Check for significance
if p_value < 0.05:
    print("There is a significant correlation between location and vital status.")
else:
    print("There is no significant correlation between location and vital status.")

# Heatmap visualization
plt.figure(figsize=(10, 8))
sns.heatmap(location_status_data, annot=True, cmap='coolwarm', fmt='d', cbar=True)
plt.title('Contingency Table Heatmap of Location vs Vital Status')
plt.xlabel('Vital Status')
plt.ylabel('Location')
plt.show()
```
